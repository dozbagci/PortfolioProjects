```python
import pandas as pd
import numpy as np
```

```python
#read the title document, put it in a dataframe
df_titles = pd.read_csv('title_basics_2018.csv')
df_titles = pd.DataFrame(df_titles)
df_titles.head()
```

```python
#inspect the dimensions of the datafreme
df_titles.shape
```

```python
#read the ratings document, put it in a dataframe and inspect the dimensions
df_ratings = pd.read_csv('title_ratings.csv')
df_ratings = pd.DataFrame(df_ratings)
df_ratings.shape
```

```python
#combine two dataframes by including only the relevant ones from the rating data and inspect the shape
df = pd.merge(df_titles, df_ratings, how='left', on='tconst')
df.shape
```

```python
df.head()
```

```python
# QUESTION 1: find how many Comedy films were made in 2018
print(df['genres'].value_counts()['Comedy'])
```

```python
# QUESTION 2: find how many films have a rating higher than 8.0
df_above_8 = df.query('averageRating >= 8.0')
df_above_8.shape
```

```python
# QUESTION 3: to find the best film of 2018, let's inspect the highest averageRating
df.sort_values('averageRating',ascending=False)
```

```python
# QUESTION 3
# above we saw that films that have the highest rating have a very low number of votes.
# this can undermine our analysis, so let's also check the highest number of votes
df.sort_values('numVotes',ascending=False)
```

```python
# QUESTION 3
# in order to implement a method that is sensitive to both rating and number of votes
# let's calculate another metric called "success" that is basically calculated by
# multiplying average rating with number of votes

df['success'] = df['averageRating']*df['numVotes']
df.sort_values('success',ascending=False)
```

```python
# QUESTION 4
# in order to find if audiences prefer longer or shorter films, I chose to do a linear regression
# Before I conduct regression analysis to predict the rating from the runtime, I need to deal with outliers
# Outlier detection is done to eliminate extraordinary and rare runtime values to contaminate our analysis
# In the end it will delete only the values 3 Standard Deviation above or below the mean value

def reject_outliers(data):
 u = np.mean(data["runtimeMinutes"])
 s = np.std(data["runtimeMinutes"])
 data_filtered = data[(data["runtimeMinutes"]>(u-3*s)) & (data["runtimeMinutes"]<(u+3*s))]
 return data_filtered

data_filtered=reject_outliers(df)
n_of_excluded_data = df.shape[0] - data_filtered.shape[0]
print(n_of_excluded_data)
```

```python
# QUESTION 4
# Implement Linear Regression
# Caution; there are NaN values which we need to exclude to be able to run the analysis.

import scipy.stats
m = ~(np.isnan(data_filtered['runtimeMinutes']) | np.isnan(data_filtered['averageRating'])) #find indexes of rows with NaNs
x_m, y_m = data_filtered['runtimeMinutes'][m], data_filtered['averageRating'][m] #exclude NaNs
result = scipy.stats.linregress(x_m,y_m) #Linear Regression
print(result)
```

```python
# QUESTION 4
# Let's plot our analysis to understand better
# There is a slight positive slope, which means longer films have slightly higher ratings
# However, this is not statistically significant

import matplotlib.pyplot as plt
plt.plot(x_m, y_m, 'o', label='original data', markersize = 3)
plt.plot(x_m, result.intercept + result.slope*x_m, 'r', label='fitted regression line')
plt.xlabel("Duration in minutes")
plt.ylabel("Average rating")
plt.legend()
plt.show()
```
