```python
# Import packages
import pandas as pd
import numpy as np

# Read in the data
df = pd.read_csv(r'C:\Users\u164590\Downloads\Wiris dataset for analyst test 2023\test_dataset.csv')
```

```python
#Check the dimensions of our data
df.shape
```

```python
#Check data types
df.dtypes
```

```python
#Convert date variables' data type to datetime
df['creation_datetime'] = pd.to_datetime(df['creation_datetime'])
df['trial_start_date'] = pd.to_datetime(df['trial_start_date'])
df['activation_date'] = pd.to_datetime(df['activation_date'])
df['date_purchased'] = pd.to_datetime(df['activation_date'])
df.dtypes
```

```python
# Drop duplicates if there is any
df.drop_duplicates()

# Check percentage of missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Dropping columns that are not necessary for our analysis

# column 'creation_date' is redundant because the date information is stored in 'creation_date_time' as well
df = df.drop('creation_date', axis = 1) # drop 'creation_date'

df.head()
```

```python
# check 'title' column further. It had 87% missing values, can be not worth to keep.

df['title'].value_counts(dropna = False)[:30]  #check the first 30 value counts
```

```python
# So many values in different alphabet characters that might create problems. Let's drop the 'title' column completely
df = df.drop('title', axis = 1) # drop 'title'
```

```python
file_name = 'test_wiris_prepped.csv'
df.to_csv(file_name)
```
